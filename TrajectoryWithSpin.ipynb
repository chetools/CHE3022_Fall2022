```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
def Cd(Re):
    a = Re/5
    b= Re/2.6e5
    c = Re/1e6

    return 24/Re + 2.6*a/(1+a**1.5) + 0.411* b**-25 / (1+b**-25) + 0.25*c/(1+c)
```

```python
Re = np.logspace(1, 7, 500)
Cd_vals = Cd(Re)
fig=make_subplots()
fig.add_scatter(x=Re, y=Cd_vals, mode='lines')
fig.update_xaxes(type="log")
fig.update_yaxes(type='log')
fig.update_layout(width=800,height=600,template='plotly_dark')
```
