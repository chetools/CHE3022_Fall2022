```python
import numpy as np
from plotly.subplots import make_subplots
```

```python
alpha=1.
dx=1.
dt=0.508
r = alpha * dt/dx**2
T0 = 20.
Tleft = 50.
Tright = 80.
```

```python
N=20
x=np.arange(N)
```

```python
T = np.full(N,T0)
T[0]=Tleft
T[-1]=Tright
fig = make_subplots()
fig.add_scatter(x=x,y=T,mode='lines', name=f'0')
for i in range(1,500):
    T[1:-1]=r*T[2:]+(1-2*r)*T[1:-1]+r*T[:-2]
    if (i % 25) == 0:
        fig.add_scatter(x=x,y=T,mode='lines',name=f'{i}')
fig.update_layout(width=800,height=600,template='plotly_dark')
```
