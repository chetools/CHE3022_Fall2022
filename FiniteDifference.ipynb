```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.linalg import solve_banded
```

```python
alpha=1.
dx=1.
dt=0.508
r = alpha * dt/dx**2
T0 = 20.
Tleft = 50.
Tright = 80.
```

```python
N=20
x=np.linspace(0,1,N)
```

```python
T = np.full(N,T0)
T[0]=Tleft
T[-1]=Tright
fig = make_subplots()
fig.add_scatter(x=x,y=T,mode='lines', name=f'0')
for i in range(1,500):
    T[1:-1]=r*T[2:]+(1-2*r)*T[1:-1]+r*T[:-2]
    if (i % 25) == 0:
        fig.add_scatter(x=x,y=T,mode='lines',name=f'{i}')
fig.update_layout(width=800,height=600,template='plotly_dark')
```

## Crank-Nicolson

```python
Ti, Tleft,Tright = 25., 50., 80.
alpha, dx, dt=1., 1., 0.2
u =  alpha*dt/2/dx**2
```

```python
N=20
x=np.linspace(0,1,N)
Abands=np.zeros((3,N-2))
Abands[0,1:]=-u
Abands[1,:]=1+2*u
Abands[2,:-1]=-u

T = np.full(N,Ti)
T[0],T[-1]=Tleft,Tright
fig = make_subplots()
fig.add_scatter(x=x,y=T,mode='lines', name=f'0')
for i in range(1,500):
    b = u*T[2:]+(1-2*u)*T[1:-1]+u*T[:-2]
    b[0]=b[0]+u*T[0]
    b[-1]=b[-1]+u*T[-1]
    T[1:-1]=solve_banded((1,1),Abands,b)
    if (i % 25) == 0:
        fig.add_scatter(x=x,y=T,mode='lines',name=f'{i*dt}')
fig.update_layout(width=800,height=600,template='plotly_dark')
```
